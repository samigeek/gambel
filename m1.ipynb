```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
import random
```

```python
def gambel(money,desir,g=5000,win_prob=0.5):
  g=5000
  all=[money]
  while (money <= desir) and (money >= 0):
    bet=random.random()
    if bet<win_prob:
      money-=g
    else:
      money+=g
    all.append(money)
  return money, all
```

```python
bet=[]
mean_bet=[]
for i in range(1000):
  money,all=gambel(100000,300000)
  bet.append(money)
  mean_bet.append(np.mean(all))
```

```python
bet=np.array(bet)
value,count=np.unique(bet,return_counts=True)
print(str(count[0]/count.sum())+' loss probality')
print(str(count[1]/count.sum())+' win probality')
```
